Import statements

```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```
import numpy as np
np.random.seed(42)

xs = 2 * np.random.rand(100, 1)
ys = 3 * xs + np.random.rand(100, 1)
zs = 2 * ys + 3 * xs + np.random.rand(100,1)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(xs,ys,zs, color="purple")
```

```
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 10, 0, 10])
fig.show()
```

```
# split the data into training and test sets
# train set
train_xs = xs[:80]
train_ys = ys[:80]
train_zs = zs[:80]
# test set
test_xs = xs[80:]
test_ys = ys[80:]
test_zs = zs[80:]
```

```
# number of epochs
epochs = 200
# learning rate
lr = .0075
```

```
# initial value for weight w1, w2 and bias b
w1 = np.random.randn(1)
w2 = np.random.randn(1)
b = np.zeros(1)
```

```
for epoch in np.arange(epochs):
  for i in np.arange(80):
    z_pred = w1 * train_xs[i] + w2 * train_ys[i] + b
    
    grad_w1 = (z_pred - train_zs[i]) * train_xs[i]
    grad_w2 = (z_pred - train_zs[i]) * train_ys[i]
    
    grad_b = (z_pred - train_zs[i])
    
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    b -= lr * grad_b
```

```
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * (w1 * test_xs[i] + w2 * test_ys[i] + b - test_zs[i]) ** 2
test_loss /= 20
```

```
pred_zs = w1 * test_xs + w2 * test_ys + b
```

```
ax = plt.axes(projection='3d')
ax.scatter3D(test_xs, test_ys, test_zs, "b.")
ax.scatter3D(test_xs, test_ys, pred_zs, "r.")
```
