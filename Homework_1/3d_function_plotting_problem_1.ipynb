3D plots are located at the bottom of the notebook

import statements 

```
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
%matplotlib inline
```

Filling array with several points in -10 to 10 range

```
x_plots = np.linspace(-10,10,100)
y_plots = np.linspace(-10,10,100)

mesh_x, mesh_y = np.meshgrid(x_plots, y_plots)
```

3D functions

```
def function_min(x,y):
  return x**2 + y**2 + 5

def function_max(x,y):
  return -1*(x**2 + y**2) + 5

def function_saddle(x,y):
  return y**2 - x**2 + 5
```

Grabbing the Z return values from the function

```
min_plots = function_min(mesh_x,mesh_y)
max_plots = function_max(mesh_x,mesh_y)
saddle_plots = function_saddle(mesh_x,mesh_y)

plt.xlim([-10,10])
plt.ylim([-10,10])
```

Printing all the points

```
print("X plots:", x_plots, '\n')
print("Y plots:", y_plots, '\n')
print("X mesh:", mesh_x, '\n')
print("Y mesh:", mesh_y,'\n')
print("Min Func z points", min_plots, '\n')
print("Max Func z points", max_plots, '\n')
print("Saddle Func z points", saddle_plots, '\n')
```

Plotting the contour and density of each function

```
plt.contour(mesh_x, mesh_y, min_plots, cmap="plasma")
```

```
plt.hist(min_plots)
```

```
plt.contour(mesh_x, mesh_y, max_plots, cmap="magma")
```

```
plt.hist(max_plots)
```

```
plt.contour(mesh_x, mesh_y, saddle_plots, cmap="viridis")
```

```
plt.hist(saddle_plots)
```

Plotting the functions in 3D space

```
fig = plt.figure()
```

```
ax = plt.axes(projection='3d')
ax.plot_wireframe(mesh_x,mesh_y,min_plots, color="purple")
```

```
ax = plt.axes(projection='3d')
ax.scatter3D(0,0,5,color="black")
ax.plot_wireframe(mesh_x,mesh_y,max_plots, color="blue")
```

```
ax = plt.axes(projection='3d')
ax.plot_wireframe(mesh_x,mesh_y, saddle_plots, color="red")
```
